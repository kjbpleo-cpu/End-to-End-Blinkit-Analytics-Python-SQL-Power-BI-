```python
#----------------BLINKIT DATASET --------#

#importing and understanding data
import pandas as pd
df=pd.read_csv('/content/blinkit_dataset.csv')
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['offer_type'].value_counts()
```

```python
#handlle missing values
df['offer_type'] = df['offer_type'].fillna('No Offer')
df['offer_type'].value_counts()
```

```python
#Feature enginerring
df['revenue'] = df['final_price'] * df['sold_quantity']
df['discount_value'] = df['price'] - df['final_price']
df['rating_category'] = pd.cut(
    df['rating'],
    bins=[0,2.5,4,5],
    labels=['Low','Average','High']
)
```

```python
#EDA
#1,Offer impact on  sales
df.groupby('offer_type')['sold_quantity'].mean().sort_values(ascending=False)
```

```python
#discountvs revenue
df.groupby('discount_pct')['revenue'].mean()
```

```python
#top category by revenue
df.groupby('category')['revenue'].sum().sort_values(ascending=False)
```

```python
#rating vs Sales
df.groupby('rating_category')['sold_quantity'].mean()
```

```python
#fast vs slow delivery
df['delivery_speed'] = df['delivery_time_min'].apply(
    lambda x: 'Fast' if x <= 30 else 'Slow'
)
df.groupby('delivery_speed')['rating'].mean()
```

```python
df.groupby('city')['revenue'].sum().sort_values(ascending=False)
```

```python
#hidden opportunity products
df[(df['rating'] > 4) & (df['sold_quantity'] < 50)][
    ['product_name','rating','sold_quantity']
]
```

```python
df.to_csv("cleaned_blinkit_data.csv", index=False)
```
